```python
import pandas as pd
```

# New Section

```python
dds = pd.read_csv("Diabetes Dataset.csv")
```

```python
dds.head()
```

```python
dds.rename(columns={"Body Mass Index)": "Body Mass Index"}, inplace=True)
```

```python
dds.head()
```

```python
dds.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dds_trData = dds.iloc[:,0:-1]
```

```python
dds_trData.head()
```

```python
y = dds.Outcome
x = dds_trData
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()

norm_data = Scaler.fit_transform(x)

print(norm_data)

x = pd.DataFrame(norm_data, columns=x.columns)
```

```python
x.head()
```

```python
# First Train the Model
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
# Now Predict the Model
y_predict_LR = model_LR.predict(x_test)
```

```python
# Find Accuracy, Precision and Recall
print(accuracy_score(y_test,y_predict_LR ))
print(classification_report(y_test,y_predict_LR))
```

```python
# Now find Confusion Matrix
confusion_matrix(y_test, y_predict_LR)
```

```python
import numpy as np
```

```python
input_data = (6, 144, 73, 189.6, 9.9, 66, 50, 99)

# Changing the input data into numpy array

input_data_into_np_arr = np.asarray(input_data)

# Reshape the array as we are predicting for one instance

input_data_reshaped = input_data_into_np_arr.reshape(1,-1)

prediction = model_LR.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print("This person is not diabetic")
else:
    print("This person is diabetic")
```
